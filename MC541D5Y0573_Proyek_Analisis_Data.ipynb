## Visualization & Explanatory Analysis

### Pertanyaan 1: Faktor apa saja yang mempengaruhi jumlah penyewaan sepeda setiap bulannya?

```python
monthly_data = day_df.groupby('mnth').agg({
    'temp': 'mean',
    'atemp': 'mean',
    'hum': 'mean',
    'windspeed': 'mean'
}).reset_index()

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.lineplot(x='mnth', y='temp', data=monthly_data, marker='o')
plt.title('Rata-rata Suhu per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Suhu (temp)')

plt.subplot(2, 2, 2)
sns.lineplot(x='mnth', y='atemp', data=monthly_data, marker='o')
plt.title('Rata-rata Suhu yang Dirasakan per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Suhu yang Dirasakan (atemp)')

plt.subplot(2, 2, 3)
sns.lineplot(x='mnth', y='hum', data=monthly_data, marker='o')
plt.title('Rata-rata Kelembaban per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Kelembaban (hum)')

plt.subplot(2, 2, 4)
sns.lineplot(x='mnth', y='windspeed', data=monthly_data, marker='o')
plt.title('Rata-rata Kecepatan Angin per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Kecepatan Angin (windspeed)')

plt.tight_layout()
plt.show()
```

Distribusi harian mendekati normal dengan puncak sekitar 4000-5000 penyewaan per hari, meskipun ada beberapa hari dengan angka lebih tinggi, kemungkinan dipengaruhi oleh faktor eksternal seperti cuaca atau akhir pekan. Sementara itu, distribusi per jam menunjukkan pola skewed ke kanan, di mana sebagian besar penyewaan terjadi pada jam-jam tertentu, mengindikasikan adanya jam sibuk. Hal ini menunjukkan bahwa pola penyewaan sepeda bervariasi tergantung skala waktu, dengan tren yang lebih stabil secara harian dibandingkan per jam.

```python
monthly_avg = day_df.groupby('mnth').agg({
    'cnt': 'mean',
    'temp': 'mean',
    'atemp': 'mean',
    'hum': 'mean',
    'windspeed': 'mean'
}).reset_index()


plt.subplot(2, 2, 1)
sns.lineplot(x='mnth', y='cnt', data=monthly_avg, marker='o')
plt.title('Rata-rata Penyewaan Sepeda per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Rata-rata Penyewaan (cnt)')

plt.subplot(2, 2, 2)
sns.lineplot(x='mnth', y='temp', data=monthly_avg, marker='o')
plt.title('Rata-rata Suhu (temp) per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Rata-rata Suhu (temp)')

plt.subplot(2, 2, 3)
sns.lineplot(x='mnth', y='atemp', data=monthly_avg, marker='o')
plt.title('Rata-rata Suhu yang Dirasakan (atemp) per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Rata-rata Suhu yang Dirasakan (atemp)')

plt.subplot(2, 2, 4)
sns.lineplot(x='mnth', y='windspeed', data=monthly_avg, marker='o')
plt.title('Rata-rata Kecepatan Angin (windspeed) per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Rata-rata Kecepatan Angin (windspeed)')

plt.tight_layout()
plt.show()
```

Berdasarkan data yang ditampilkan, jumlah penyewaan sepeda per bulan menunjukkan pola yang mirip dengan tren suhu dan suhu yang dirasakan, dengan peningkatan yang signifikan dari awal tahun hingga puncaknya di bulan Juni hingga September sebelum menurun kembali di akhir tahun. Hal ini mengindikasikan bahwa suhu yang lebih tinggi berkontribusi pada meningkatnya penyewaan sepeda, kemungkinan karena cuaca yang lebih nyaman untuk bersepeda. Sebaliknya, kecepatan angin yang lebih tinggi di awal tahun tampaknya berkorelasi dengan jumlah penyewaan yang lebih rendah, mungkin karena kondisi angin kencang kurang ideal untuk bersepeda. Penurunan penyewaan setelah bulan September dapat disebabkan oleh faktor cuaca yang mulai lebih dingin atau kondisi lingkungan lainnya yang kurang mendukung aktivitas luar ruangan. Analisis ini menunjukkan bahwa faktor iklim, terutama suhu dan kecepatan angin, memiliki pengaruh besar terhadap tren penyewaan sepeda sepanjang tahun.

### Pertanyaan 2: Bagaimana pola penyewaan sepeda berdasarkan hari dan jam ?

```python
weekday_avg = day_df.groupby('weekday')['cnt'].mean().reset_index()


plt.figure(figsize=(10, 5))
sns.lineplot(x='weekday', y='cnt', data=weekday_avg, marker='o')
plt.title('Rata-rata Penyewaan Sepeda Berdasarkan Hari dalam Seminggu')
plt.xlabel('Hari (0=Senin, 6=Minggu)')
plt.ylabel('Rata-rata Jumlah Penyewaan')
plt.xticks(ticks=[0, 1, 2, 3, 4, 5, 6], labels=['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'])
plt.show()
```

Berdasarkan grafik yang menunjukkan rata-rata penyewaan sepeda per hari dalam seminggu, terlihat bahwa jumlah penyewaan meningkat secara bertahap dari Senin hingga mencapai puncaknya pada hari Jumat dan Sabtu, kemudian sedikit menurun pada hari Minggu. Pola ini mengindikasikan bahwa penyewaan sepeda cenderung lebih tinggi menjelang akhir pekan, kemungkinan karena meningkatnya aktivitas rekreasi dan penggunaan sepeda untuk bersantai.

Untuk menganalisis pola penyewaan berdasarkan jam, diperlukan data lebih lanjut yang menunjukkan tren harian. Namun, biasanya dalam analisis serupa, penyewaan sepeda cenderung meningkat pada pagi hari (sekitar jam 7-9) saat orang pergi bekerja atau sekolah, lalu menurun di siang hari, dan kembali meningkat di sore hingga malam hari (sekitar jam 16-19) saat orang pulang kerja atau beraktivitas rekreasi. Jika ada data jam, kita dapat mengonfirmasi apakah pola ini berlaku pada dataset yang sedang dianalisis.

```python
hour_avg = hour_df.groupby('hr')['cnt'].mean().reset_index()

plt.figure(figsize=(12, 5))
sns.lineplot(x='hr', y='cnt', data=hour_avg, marker='o')
plt.title('Rata-rata Penyewaan Sepeda Berdasarkan Jam dalam Sehari')
plt.xlabel('Jam')
plt.ylabel('Rata-rata Jumlah Penyewaan')
plt.xticks(ticks=range(24), labels=[f'{i}:00' for i in range(24)], rotation=45)
plt.grid(True)
plt.show()
```

Pola penyewaan sepeda dalam sehari menunjukkan dua puncak utama yang terjadi pada pagi dan sore hari. Pada dini hari hingga menjelang subuh (00:00 - 05:00), jumlah penyewaan sangat rendah, kemudian mulai meningkat secara signifikan sekitar pukul 07:00 - 09:00. Lonjakan ini kemungkinan besar terjadi karena banyak orang menggunakan sepeda untuk keperluan berangkat kerja atau sekolah. Setelah itu, jumlah penyewaan mengalami sedikit penurunan tetapi tetap stabil sepanjang siang hingga sore hari. Puncak penyewaan tertinggi terjadi pada pukul 17:00 - 18:00, yang dapat dikaitkan dengan aktivitas pulang kerja atau waktu luang di sore hari. Setelah pukul 19:00, jumlah penyewaan mulai menurun secara bertahap hingga tengah malam. Pola ini menunjukkan bahwa sepeda banyak digunakan sebagai alat transportasi utama di pagi dan sore hari, sementara pada waktu lainnya penggunaannya lebih rendah.
